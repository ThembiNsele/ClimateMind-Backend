```python
import pandas as pd 

data = pd.read_csv("pvq21CENTRED.csv", sep=',')
lib_data = data.copy()
con_data = data.copy()
```

```python
con_data.head(20)
```

```python
lrscale = lib_data['lrscale']
lrscale.hist(bins=10, figsize=[14,6])
```

```python
def normalise (value, lib_con = None):

    if lib_con == 'l':

        if value < 3:
            return 0 #moderate
        else:
            return 1 #conservative
    elif lib_con == 'r':
        
        if value > 7:
            return 1 #conservative
        else:
            return 0 #moderate
```

```python
lib_data['lrscale'] = lib_data['lrscale'].apply(normalise, args = ('l'))
con_data['lrscale'] = con_data['lrscale'].apply(normalise, args = ('r'))
data['lrscale'] = data['lrscale'].apply(normalise, args = ('l'))
```

## Conservative Data Splitting

```python
X_train_con, X_test_con, y_train_con, y_test_con = train_test_split(X_con, y_con, test_size=0.20, random_state=10)
```

## LASSO (≈ 0.318)

```python
model = Lasso(alpha=1.0)
# define model evaluation method
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
# evaluate model
scores = cross_val_score(model, X_train_lib, y_train_lib, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
# force scores to be positive|
scores = np.absolute(scores)
print('Mean MAE: %.3f (%.3f)' % (np.mean(scores), np.std(scores)))
```

```python
y_test_lib
```

```python
model.fit(X_train_lib, y_train_lib)
lasso_predicted = model.predict(X_test_lib)
print(lasso_predicted.astype(int))
print("Accuracy of: ", np.mean(lasso_predicted.astype(int) == y_test_lib.astype(int)))
```
